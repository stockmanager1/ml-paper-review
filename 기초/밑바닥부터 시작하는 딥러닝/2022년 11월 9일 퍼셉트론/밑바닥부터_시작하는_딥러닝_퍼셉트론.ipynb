#퍼셉트론 구현하기

##and 게이트 구현하기

```python
def AND(x1,x2):
  w1,w2,theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp < theta:
    return 0
  else:
    return 0
```

#가중치와 편향 구현하기

```python
def AND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  b = -0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
AND(1,1)
```

편향의 값은 뉴런이 얼마나 쉽게 활성화되는지를 결정한다.

```python
def NAND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([-0.5,-0.5])
  b = 0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
def OR(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  b = -0.2
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

AND와 편향과 가중치의 부호가 반대다.
