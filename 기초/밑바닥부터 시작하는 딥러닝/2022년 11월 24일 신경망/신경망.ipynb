#5. 출력층 설계하기.

일반적으로 회귀분석에는 항등함수를, 분류에는 소프트맥스 함수를 사용합니다.

##5.1 항등함수와 소프트 맥스 함수 구현하기

image.png

분자는 지수함수를 말하고, 분모는 모든 입력신호의 지수함수의 합을 의미하므로 항등함수와 다르게 소프트맥스의 출력은 모든 입력신호로부터 화살표를 받습니다.

image.png

소프트 맥스함수를 구현해보자.

```python
a = np.array([0.3,2.9,4.0])

exp_a = np.exp(a)

print(exp_a)
```

```python
sum_exp_a = np.sum(exp_a)
print(sum_exp_a)
```

```python
y = exp_a / sum_exp_a
print(y)
```

```python
def softmax(a):
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a

  return y 
```

##5.2 소프트맥스 함수 구현 시 문제점



소프트 맥스 함수는 오버플로 문제(컴퓨터 능력을 벗어나는 연산)가 있다. 

image.png

따라서 이를 해결하기 위해 c라는 임의의 정수를 분자와 분모 양쪽에 곱했다.

```python
def softmax(a):
  c = np.max(a)
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a

  return y 
```

##5.3 소프트맥스 함수의 특징

```python
a = np.array([0.3,2.9,4.0])
y = softmax(a)
print(y)

print(np,sum(y))
```

소프트 맥스 함수는 출력이 0에서 1.0사이에 존재한다. 또 소프트맥스 함수 출력의 총합은 1이다. 이걸로 우리는 각 경우에 대한 확률을 구할 수 있는데, 여기서 높은 확률값이 신경망은 값으로 인식한다.
